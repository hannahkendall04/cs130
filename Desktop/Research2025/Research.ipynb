## ATAC counts shape

```python
import h5py
from scipy.sparse import csr_matrix

# Load the HDF5 file
with h5py.File("atac_window_20k.h5", "r") as f:
    group = f["matrix"]
    data = group["data"][:]
    indices = group["indices"][:]
    indptr = group["indptr"][:]
    raw_shape = tuple(group["shape"][:])  # (features, cells)

# Correct the shape: rows = len(indptr) - 1, columns = raw_shape[0]
corrected_shape = (len(indptr) - 1, raw_shape[0])  # (7941, 160724)

# Construct the sparse matrix
atac_matrix = csr_matrix((data, indices, indptr), shape=corrected_shape)

# Print the shape
print("Matrix shape:", atac_matrix.shape)

# Print the first 5 rows and 5 columns
print("Matrix head (5x5):")
print(atac_matrix[:10, :50].toarray())
```
